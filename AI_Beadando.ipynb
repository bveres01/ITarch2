```python
# Könyvtárak importálása
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Adatok beolvasása
cases_df = pd.read_csv("covid-cases.csv")  # COVID-esetszám adatok
pop_df = pd.read_csv("world-population.csv")  # Népességi adatok

# Csak a szükséges oszlopokat tartjuk meg és átnevezzük
pop_df = pop_df[['Name', '2020']].rename(columns={'Name': 'Country', '2020': 'Population_2020'})

# A két adatot összeolvasztjuk ország alapján
merged_df = pd.merge(cases_df, pop_df, on='Country', how='left')

# Eldobjuk azokat a sorokat, ahol nincs népességi adat
merged_df = merged_df.dropna(subset=['Population_2020'])

# Átváltás: esetszám és halálozás 100 ezer főre vetítve
merged_df['Cases_per_100k'] = merged_df['New_cases'] / (merged_df['Population_2020'] * 1000) * 100000
merged_df['Deaths_per_100k'] = merged_df['New_deaths'] / (merged_df['Population_2020'] * 1000) * 100000
```
